# CORONA TREND IN MARCH



```python
#importing relevant libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.get_backend()
%matplotlib notebook
```

```python
#Importing datasets from csv files
df2= pd.read_csv('CoronaMI.csv')
df3=pd.read_csv('CoronaCalifornia.csv')

#Making Date in a relevant Format
df2['Date'] = df2['dateChecked'].apply(lambda x: x[:10])
df3['Date'] = df3['dateChecked'].apply(lambda x: x[:10])

#Grouping date to plot time series(Michigan)
df2_positive = df2.groupby('Date')['positive'].agg(sum)
df2_death = df2.groupby('Date')['death'].agg(sum)

#Grouping date to plot time series(California)
df3_positive = df3.groupby('Date')['positive'].agg(sum)
df3_death = df3.groupby('Date')['death'].agg(sum)

#Plotting the Michigan Dataset
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(df2_positive,'b', alpha = 1, label='Positive Cases in Michigan')
plt.plot(df2_death,'b',alpha=0.4,label='Deaths in Michigan')
plt.xlabel('March Dates')
plt.ylabel('Number of Cases')
plt.title('Trendline of Coronavirus in Michigan State for March')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xticks([2, 6, 10 ,14, 18],['04','08','12','16','20'])
plt.gca().fill_between(range(len(df2)),df2_positive,df2_death,facecolor='blue',alpha=0.15)
plt.legend(frameon=False)

#Plotting the Chicago Dataset
plt.subplot(1,2,2)
plt.plot(df3_positive,'r', alpha = 1, label='Positive Cases in Chicago')
plt.plot(df3_death,'r',alpha=0.4,label='Deaths in Chicago')
plt.legend(frameon=False)
plt.xlabel('March Dates')
plt.ylabel('Number of Cases')
plt.title('Trendline of Coronavirus in California State for March')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().fill_between(range(len(df3)),df3_positive,df3_death,alpha=0.15,facecolor='red')
plt.xticks([2, 6, 10 ,14, 18],['04','08','12','16','20'])

plt.tight_layout()
```
